```python
#import necessary libraries
import pandas as pd
```

```python
UAL_info = pd.read_csv("Data/UAL_Fundamentals.csv",index_col=[2,3]).sort_index()
DAL_info = pd.read_csv("Data/DAL_Fundamentals.csv",index_col=[2,3]).sort_index()
LUV_info = pd.read_csv("Data/LUV_Fundamentals.csv",index_col=[2,3]).sort_index()
```

```python
DAL_info.index.max(),LUV_info.index.max(),UAL_info.index.max()
```

```python
def Gen_Bar_Chart(data:pd.DataFrame, yval='Revenue', title='Revenue',rot=90):
    data.plot.bar(y=yval, title=title,figsize=(15,5), rot=90)
def Gen_Line_Chart(data:pd.DataFrame, yval='Revenue', title='Revenue',rot=90):
    data.plot.line(y=yval, title=title, figsize=(15,5), rot=90)
```

```python
exclueded_cols = ['Unnamed: 0','Ticker','Fiscal Period', 'Revenue','Net Income', 'Gross Profit Margin', 'Net Profit Margin', 'Operating Margin',  'EPS', 'Cash & Cash Equivalents',  'Total Equity','Quick Ratio', 'ROE', 'Stock Price by Quarter', 'PE and PS']
 
ticker_info =[UAL_info, LUV_info, DAL_info]

charts=[]
for stk_sym in ticker_info:
    columns_2_plot = [col for col in stk_sym.columns if col not in exclueded_cols]
    print(len(stk_sym))
    for col in columns_2_plot:
        print(f'Working on column {col}')
        chart_pair = Gen_Bar_Chart(stk_sym,yval=col,title=f"{col} by Quarter over the last 10 years")
        #(Gen_Bar_Chart(stk_sym,yval=col,title=f"{col} by Quarter over the last 10 years"), 
        #             Gen_Line_Chart(stk_sym,yval=col,title=f"{col} by Quarter over the last 10 years"))
        charts.append(chart_pair)
```
