```python
# Created Function for plots for Passenger Data, Flights, Passenger-Miles, Seat-Miles, and Load data

# Passenger Data

def us_chart1():
    bar_plot = us_passengers_final.hvplot.bar(label= "All U.S. Carriers - All Airports", ylabel = "Number of Passengers", color= "Orange", shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

def dal_chart1():
    bar_plot = dal_passengers_final.hvplot.bar(label= "Delta Airlines - All Airports", ylabel = "Number of Passengers", color="Green", shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

def luv_chart1():
    bar_plot = luv_passengers_final.hvplot.bar(label= "Southwest Airlines - All Airports", ylabel = "Number of Passengers", color="Purple",shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

def ual_chart1():
    bar_plot = ual_passengers_final.hvplot.bar(label= "United Airlines - All Airports", ylabel = "Number of Passengers", color="Yellow", shared_axes= False  ,width=900,height=500).opts(yformatter="%.0f")
    return bar_plot


# Flights Data

def us_chart2():
    bar_plot = us_flights_final.hvplot.bar(label= "All U.S. Carriers Flights - All Airports", ylabel = "Number of Flights", color="Orange", shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

def dal_chart2():
    bar_plot = dal_flights_final.hvplot.bar(label= "Delta Airlines Flights - All Airports", ylabel = "Number of Flights", color="Green", shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

def luv_chart2():
    bar_plot = luv_flights_final.hvplot.bar(label= "Southwest Airlines Flights - All Airports", ylabel = "Number of Flights", color="Purple", shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

def ual_chart2():
    bar_plot = ual_flights_final.hvplot.bar(label= "United Airlines Flights - All Airports", ylabel = "Number of Flights", color="Yellow", shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

# Passenger-Miles Data

def us_chart3():
    bar_plot = us_passenger_miles_final.hvplot.bar(label= "All U.S. Carriers Airlines Revenue Passenger-Miles - All Airports" , ylabel = "(The number of passengers and the distance flown in thousands (000)) ", color="Orange", shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

def dal_chart3():
    bar_plot = dal_passenger_miles_final.hvplot.bar(label= "Delta Airlines Revenue Passenger-Miles - All Airports", ylabel = "(The number of passengers and the distance flown in thousands (000)) ", color="Green", shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

def luv_chart3():
    bar_plot = luv_passenger_miles_final.hvplot.bar(label= "Southwest Airlines Revenue Passenger-Miles - All Airports", ylabel = "(The number of passengers and the distance flown in thousands (000)) ", color="Purple", shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

def ual_chart3():
    bar_plot = ual_passenger_miles_final.hvplot.bar(label= "United Airlines Revenue Passenger-Miles - All Airports", ylabel = "(The number of passengers and the distance flown in thousands (000)) ", color="Purple", shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

# Seat-Miles Data

def us_chart4():
    bar_plot = us_seat_miles_final.hvplot.bar(label= "All U.S. Carriers Available Seat-Miles - All Airports" , ylabel = "(the number of seats and the distance flown in thousands (000)) ", color="Orange", shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

def dal_chart4():
    bar_plot = dal_seat_miles_final.hvplot.bar(label= "Delta Airlines Available Seat-Miles - All Airports", ylabel = "(the number of seats and the distance flown in thousands (000)) ", color="Green", shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

def luv_chart4():
    bar_plot = luv_seat_miles_final.hvplot.bar(label= "Southwest Airlines Available Seat-Miles - All Airports", ylabel = "(the number of seats and the distance flown in thousands (000)) ", color="Purple", shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

def ual_chart4():
    bar_plot = ual_seat_miles_final.hvplot.bar(label= "United Airlines Available Seat-Miles - All Airports", ylabel = "(the number of seats and the distance flown in thousands (000)) ", color="Purple", shared_axes= False , width=900,height=500).opts(yformatter="%.0f")
    return bar_plot

# Load Factor Data

def us_chart5():
    bar_plot = us_load_final.hvplot.bar(label= "All U.S. Carriers Load Factor - All Airports" , ylabel = "(passenger-miles as a proportion of available seat-miles in percent (%)) ", color="Orange", shared_axes= False , width=900,height=500)
    return bar_plot

def dal_chart5():
    bar_plot = dal_load_final.hvplot.bar(label= "Delta Airlines Load Factor - All Airports", ylabel = "(passenger-miles as a proportion of available seat-miles in percent (%)) ", color="Green", shared_axes= False , width=900,height=500)
    return bar_plot

def luv_chart5():
    bar_plot = luv_load_final.hvplot.bar(label= "Southwest Airlines Load Factor - All Airports", ylabel = "(passenger-miles as a proportion of available seat-miles in percent (%)) ", color="Purple", shared_axes= False , width=900,height=500)
    return bar_plot

def ual_chart5():
    bar_plot = ual_load_final.hvplot.bar(label= "United Airlines Load Factor  - All Airports", ylabel = "(passenger-miles as a proportion of available seat-miles in percent (%)) ", color="Purple", shared_axes= False , width=900,height=500)
    return bar_plot
```

```python
# Created variables for dashboard

dashboard_title = " Bureau of Transportation Statistics - Aviation Data"

dashboard_welcome = " The dashboard presents a visual analysis of aviation data from All U.S. Carriers, Delta Airlines, Southwest Airlines, and United Airlines"

dashboard_page = pn.Column(dashboard_title, dashboard_welcome)

passenger_data = pn.Column(pn.Pane(us_chart1()), pn.Pane(dal_chart1()), pn.Pane(luv_chart1()), pn.Pane(ual_chart1()))

flights_data = pn.Column(pn.Pane(us_chart2()), pn.Pane(dal_chart2()), pn.Pane(luv_chart2()), pn.Pane(ual_chart2()))

passenger_miles_data = pn.Column(pn.Pane(us_chart3()), pn.Pane(dal_chart3()), pn.Pane(luv_chart3()), pn.Pane(ual_chart3()))

seat_miles_data = pn.Column(pn.Pane(us_chart4()), pn.Pane(dal_chart4()), pn.Pane(luv_chart4()), pn.Pane(ual_chart4()))

load_data = pn.Column(pn.Pane(us_chart5()), pn.Pane(dal_chart5()), pn.Pane(luv_chart5()), pn.Pane(ual_chart5()))


#Create Dashboard
dashboard_tabs = pn.Tabs(            
                ("Passenger Data", passenger_data),
                ("Flight Data", flights_data),
                ("Passenger-Miles Data", passenger_miles_data),
                ("Available Seat-Miles", seat_miles_data),
                ("Load Data", load_data)
)
                
```
