# **Valuation of US Airline Industry**

## **Gathering Stock Information for Southwest Airlines (LUV), Delta Airlines (DEL) and United Airlines (UAL)**

```python
#import necessary libraries
import os
import sys
import requests
from dotenv import load_dotenv
import pandas as pd
import csv as csv
import alpaca_trade_api as tradeapi
```

```python
if load_dotenv():
    print('The environment variables have been loaded')
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")
if type(alpaca_secret_key) == str:
    print('The Alpaca Key has been loaded')
```

```python
# Create the Alpaca API object
api=tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2"
)
```

```python
tickers=["LUV","DAL","UAL"]
timeframe="1D"
```

```python
# Get stock price data from 2011-01-01 to 2021-06-25
start_date1=pd.Timestamp("2011-01-01",tz="America/New_York").isoformat()
end_date1=pd.Timestamp("2014-01-01",tz="America/New_York").isoformat()
```

```python
df_stock_data1=api.get_barset(
        tickers,
        timeframe,
        start=start_date1,
        end=end_date1,
        limit=1000

).df
```

```python
df_stock_data1.tail()
```

```python
start_date2=pd.Timestamp("2014-01-01",tz="America/New_York").isoformat()
end_date2=pd.Timestamp("2018-01-01",tz="America/New_York").isoformat()

df_stock_data2=api.get_barset(
        tickers,
        timeframe,
        start=start_date2,
        end=end_date2,
        limit=1000

).df
df_stock_data2.tail()
```

```python
start_date3=pd.Timestamp("2018-01-01",tz="America/New_York").isoformat()
end_date3=pd.Timestamp("2021-06-25",tz="America/New_York").isoformat()

df_stock_data3=api.get_barset(
        tickers,
        timeframe,
        start=start_date3,
        end=end_date3,
        limit=1000

).df

df_stock_data3.tail()
```

```python
airline_stock=pd.concat([df_stock_data1,df_stock_data2,df_stock_data3])
airline_stock.head()
```

```python
airline_stock.reset_index(inplace=True)
airline_stock['time'] = pd.to_datetime(airline_stock['time']).dt.date
airline_stock.set_index('time',inplace=True)
```

## **Gathering the fundamental data for the stocks**

```python
import pandas as pd
import simfin as sf
import numpy as np
%matplotlib inline
```

```python
sf.set_data_dir('~/Data/')      # Changed from '~/Data/' Data in the home directory
```

```python
sf.set_api_key(api_key='free')
```

```python
%%time
market = 'us'
df_income = sf.load_income(variant='quarterly', market=market)
df_bal_sheet =sf.load_balance(variant='quarterly', market=market)
```

```python
#Filtering the fundamental data for stocks of interest
tickers = ["LUV","DAL","UAL"]
airline_income_df=df_income.loc[tickers].reset_index()
airline_balsheet_df=df_bal_sheet.loc[tickers].reset_index()
```

```python
airline_income_df.head()
```

```python
airline_balsheet_df.head()
```

```python
#Calculating the Balance Sheet Ratios

airline_balsheet_df['Quick Ratio']=(airline_balsheet_df['Total Current Assets']-airline_balsheet_df['Inventories'])/airline_balsheet_df['Total Current Liabilities']
airline_balsheet_df['Current Ratio']=airline_balsheet_df['Total Current Assets']/airline_balsheet_df['Total Current Liabilities']
airline_balsheet_df['Debt-to-Equity Ratio']=airline_balsheet_df['Total Liabilities']/airline_balsheet_df['Total Equity']
airline_balsheet_df['Long Term Debt-to-Equity Ratio']=airline_balsheet_df['Long Term Debt']/airline_balsheet_df['Total Equity']
airline_balsheet_df['Short Term Debt-to-Equity Ratio']=airline_balsheet_df['Short Term Debt']/airline_balsheet_df['Total Equity']
```

```python
#calculating income statement ratios

airline_income_df["Gross Profit Margin"]=airline_income_df['Gross Profit']/airline_income_df['Revenue']
airline_income_df["Net Profit Margin"]=airline_income_df['Net Income']/airline_income_df['Revenue']
airline_income_df["Operating Margin"]=airline_income_df['Operating Income (Loss)']/airline_income_df['Revenue']
airline_income_df['EPS']=airline_income_df['Net Income']/airline_income_df['Shares (Basic)']
airline_income_df['Interest Coverage Ratio']=airline_income_df['Net Income']/airline_income_df['Shares (Basic)']
```

```python
inc_columns=['Ticker','Report Date','Fiscal Year','Fiscal Period','Shares (Basic)',
             'Revenue','Net Income','Gross Profit Margin','Net Profit Margin',
             'Operating Margin', 'EPS','Interest Coverage Ratio']
bal_columns=['Cash, Cash Equivalents & Short Term Investments','Total Assets','Total Equity',
             'Quick Ratio', 'Current Ratio', 'Debt-to-Equity Ratio','Long Term Debt-to-Equity Ratio',
             'Short Term Debt-to-Equity Ratio']
```

```python
df_airline_ratios = pd.concat([airline_income_df[inc_columns],airline_balsheet_df[bal_columns]],join='inner',axis='columns')#, keys=['Ticker','Qtr'])
df_airline_ratios.head()
```

## **Combing the stock price data with Fundamental data to calculate valuation ratios**

```python
# Filtering the airline stock price data for closing prices
LUV_stock=airline_stock['LUV']['close']
DAL_stock=airline_stock['DAL']['close']
UAL_stock=airline_stock['UAL']['close']
```

```python
# Concatenating the LUV stock prices to the ratios
LUV_fundamentals =df_airline_ratios.loc[df_airline_ratios['Ticker']=='LUV']
LUV_fundamentals.set_index('Report Date',inplace=True)
LUV_data=pd.concat([LUV_fundamentals,LUV_stock],axis='columns',join='inner')
LUV_data.rename(columns={"close":"LUv Price"},inplace=True)

#Calculating P/E ,P/S and P/B ratios for Southwest Airlines
LUV_data["PE Ratio"]=LUV_data["LUv Price"]/LUV_data["EPS"]
LUV_data["PS Ratio"]=(LUV_data["LUv Price"]*LUV_data["Shares (Basic)"])/LUV_data["Revenue"]
LUV_data["PB Ratio"]=(LUV_data["LUv Price"]*LUV_data["Shares (Basic)"])/LUV_data["Total Equity"]
```

```python
# Concatenating the DAL stock prices to the ratios
DAL_fundamentals =df_airline_ratios.loc[df_airline_ratios['Ticker']=='DAL']
DAL_fundamentals.set_index('Report Date',inplace=True)
DAL_data=pd.concat([DAL_fundamentals,LUV_stock],axis='columns',join='inner')
DAL_data.rename(columns={"close":"DAL Price"},inplace=True)

#Calculating P/E ,P/S and P/B ratios for Delta Airlines
DAL_data["PE Ratio"]=DAL_data["DAL Price"]/DAL_data["EPS"]
DAL_data["PS Ratio"]=(DAL_data["DAL Price"]*DAL_data["Shares (Basic)"])/DAL_data["Revenue"]
DAL_data["PB Ratio"]=(DAL_data["DAL Price"]*DAL_data["Shares (Basic)"])/DAL_data["Total Equity"]
```

```python
UAL_fundamentals =df_airline_ratios.loc[df_airline_ratios['Ticker']=='UAL']
UAL_fundamentals.set_index('Report Date',inplace=True)
UAL_data=pd.concat([DAL_fundamentals,LUV_stock],axis='columns',join='inner')
UAL_data.rename(columns={"close":"UAL Price"},inplace=True)

#Calculating P/E ,P/S and P/B ratios for United Airlines
UAL_data["PE Ratio"]=UAL_data["UAL Price"]/UAL_data["EPS"]
UAL_data["PS Ratio"]=(UAL_data["UAL Price"]*UAL_data["Shares (Basic)"])/UAL_data["Revenue"]
UAL_data["PB Ratio"]=(UAL_data["UAL Price"]*UAL_data["Shares (Basic)"])/UAL_data["Total Equity"]
```
